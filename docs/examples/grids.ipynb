# Working with Grids


This example demonstrates how to create and work with different grid types in earthkit-geo. The `earthkit.geo.grids` package supports several common grid types used in meteorology and climate science.

```python
from earthkit.geo import grids
```

## Creating grids

earthkit-geo provides high-level object oriented grid definitions for a range of common grid types.

### Equirectangular

The most simple grids are probably regular latitude-longitude (or "equirectangular") grids. We can create a 5° × 5° grid like this:

```python
grid = grids.Equirectangular(5, 5)

# Plot the grid cell centroids
grid.plot()
```

earthkit-geo provides useful methods on a grid to extract things like its shape, total number of points, latitude and longitude values and bounding box.

```python
grid.shape
```

```python
grid.size()
```

```python
grid.to_latlon()
```

```python
grid.bounding_box()
```

### HEALPix

A HEALPix (Hierarchical Equal Area isoLatitude Pixelization) grid divides the sphere into equal-area pixels arranged in iso-latitude rings. It is commonly used in astrophysics and climate science for its equal-area property and efficient hierarchical structure.

HEALPix grids are defined by `nside` (which defines the number of pixels with the formula $12 × nside^2$) and the ordering scheme (ring or nested):

```python
healpix = grids.HEALPix(nside=32, order="ring")
healpix.plot()
```

## Creating grids from a grid spec dictionary

You can also instantiate grids from a grid spec dictionary using the `Grid.from_dict()` method. For example, to create an Octahedral O64 grid:

```python
O64 = grids.Grid.from_dict({"grid": "O64"})
O64.plot()
```
